```python
import time
import random

# ---------------- SORTING ALGORITHMS ---------------- #

def bubble_sort(arr):
    a = arr.copy()
    n = len(a)
    for i in range(n):
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a

def insertion_sort(arr):
    a = arr.copy()
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a

def selection_sort(arr):
    a = arr.copy()
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# ---------------- PERFORMANCE TEST ---------------- #

def measure_time(sort_func, data):
    start = time.time()
    sort_func(data)
    return time.time() - start

def analyze_dataset(data):
    algorithms = {
        "Bubble Sort": bubble_sort,
        "Insertion Sort": insertion_sort,
        "Selection Sort": selection_sort,
        "Merge Sort": merge_sort,
        "Quick Sort": quick_sort
    }

    results = {}

    for name, func in algorithms.items():
        t = measure_time(func, data)
        results[name] = t

    # Sort results by time
    sorted_results = sorted(results.items(), key=lambda x: x[1])

    print("\n--- Performance Results ---")
    for algo, t in sorted_results:
        print(f"{algo}: {t:.6f} sec")

    best_algo = sorted_results[0][0]
    print(f"\n✅ Best Algorithm for this dataset: {best_algo}")

    return best_algo

# ---------------- DATASET GENERATOR ---------------- #

def generate_dataset(size, type="random"):
    if type == "sorted":
        return list(range(size))
    elif type == "reversed":
        return list(range(size, 0, -1))
    else:
        return [random.randint(1, 10000) for _ in range(size)]

# ---------------- MAIN ---------------- #

if __name__ == "__main__":
    size = 1000

    print("Dataset Type: Random")
    data = generate_dataset(size, "random")
    analyze_dataset(data)

    print("\nDataset Type: Sorted")
    data = generate_dataset(size, "sorted")
    analyze_dataset(data)

    print("\nDataset Type: Reversed")
    data = generate_dataset(size, "reversed")
    analyze_dataset(data)
```
